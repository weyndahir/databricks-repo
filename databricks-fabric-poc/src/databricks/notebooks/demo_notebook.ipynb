# Databricks POC Demo Notebook

This notebook demonstrates the capabilities of Databricks for data processing and analysis. It includes examples of data ingestion, transformation, and visualization using Databricks features.

```python
# Import necessary libraries
from pyspark.sql import SparkSession
import pandas as pd

# Initialize Spark session
spark = SparkSession.builder.appName("Databricks POC").getOrCreate()

# Load sample data
data = spark.read.csv("dbfs:/databricks-datasets/COVID/CORD-19/2020-03-13/all_sources_metadata_2020-03-13.csv", header=True, inferSchema=True)

# Show the data
data.show()
data.printSchema()
```

## Data Transformation

In this section, we will perform some basic data transformations using Spark DataFrame operations.

```python
# Data transformation example: filtering data
filtered_data = data.filter(data['publish_time'] > 0)

# Show filtered data
filtered_data.show()
```

## Data Visualization

This section demonstrates how to visualize data using Databricks built-in visualization tools.
